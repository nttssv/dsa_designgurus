 Time complexity: measure how long an algorithm takes to run based on the size of the input

Space complexity: measure of how uch memory an algorithm requires based on the size of the input

Big-O = expressing the time (or space) complexity of an algorithm


| Big-O | Name   | Description | Example |
|--------|--------|--------------|----------------|
| O(1)   | Constant | Execution remains unchanged irrespective of input data        | checking if the stack is empty           |
| O(n) | Linear |The running time increases linearly with the size of the input | Linear traversal of a list |
|O(log n) | Lograithimic | Increases by one unit for each doubling of input data | Finding an item in a balanced search tree|
| O(n^2)| Quadratic.|proportional to the square of the number of elements. | Checking all possible pairs in an array|
|O(n^3) |Cubic| Propotional to the cube of the number of elements |   Matrix Multiplication|
| O(2^n) | Exponential time| Time doubles for every new element added | Generating all subsets of a given set
| O(n!) |Factorial time | Complexity grows factorially based on the size of the dataset | Determining all permutations of a given list.










```python
import time
import tracemalloc
import matplotlib.pyplot as plt
import random

# Linear Search: O(n) time complexity
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

def measure_performance():
    sizes = [10, 100, 1_000, 10_000, 100_000, 200_000]
    times = []
    memories = []

    for size in sizes:
        arr = list(range(size))  # Sorted array
        x = -1  # Worst case: not in array

        tracemalloc.start()
        start_time = time.time()

        linear_search(arr, x)

        end_time = time.time()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        times.append(end_time - start_time)
        memories.append(peak / 1024)  # in KB

    return sizes, times, memories

def plot_performance(sizes, times, memories):
    plt.figure(figsize=(12, 5))

    # Time Plot
    plt.subplot(1, 2, 1)
    plt.plot(sizes, times, marker='o')
    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Linear Search Runtime vs Array Size')

    # Memory Plot
    plt.subplot(1, 2, 2)
    plt.plot(sizes, memories, marker='o', color='orange')
    plt.xlabel('Array Size')
    plt.ylabel('Memory (KB)')
    plt.title('Linear Search Memory Usage vs Array Size')

    plt.tight_layout()
    plt.show()

def main():
    print("Demonstrating Linear Search (Big-O: O(n))")
    arr = [3, 5, 7, 9, 11]
    x = 7
    result = linear_search(arr, x)
    print(f"Index of {x}:", result)

    sizes, times, memories = measure_performance()
    plot_performance(sizes, times, memories)

if __name__ == "__main__":
    main()
```

```python
import time
import matplotlib.pyplot as plt
import tracemalloc
import random

# Binary search implementation
def binary_search(arr, x):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1

    return -1

# Main function to test time and memory
def test_binary_search():
    print("Binary Search Time Complexity: O(log n)")

    sizes = [10, 100, 1000, 5000, 10000, 50000, 100000]
    times = []
    memories = []

    for size in sizes:
        arr = sorted(random.sample(range(size * 2), size))
        x = random.choice(arr)

        tracemalloc.start()
        start_time = time.perf_counter()
        result = binary_search(arr, x)
        end_time = time.perf_counter()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        times.append((end_time - start_time) * 1000)  # in ms
        memories.append(peak / 1024)  # in KB

        print(f"Size: {size:>6}, Time: {times[-1]:.6f} ms, Memory: {memories[-1]:.2f} KB, Found index: {result}")

    # Plot time
    plt.figure()
    plt.plot(sizes, times, marker='o')
    plt.title('Binary Search Runtime vs Input Size')
    plt.xlabel('Array Size')
    plt.ylabel('Time (ms)')
    plt.grid(True)
    plt.show()

    # Plot memory
    plt.figure()
    plt.plot(sizes, memories, marker='o')
    plt.title('Binary Search Memory Usage vs Input Size')
    plt.xlabel('Array Size')
    plt.ylabel('Memory (KB)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    test_binary_search()
```

 Big-O Explanation:

		Time Complexity: O(log n) for binary search (halves the search space each time).

		Space Complexity: O(1) (no additional memory used aside from input list).

 Explanation:

		Bubble Sort repeatedly steps through the list, compares adjacent elements and swaps them if they are in the wrong order.
		Its inefficient for large lists  useful mostly for educational purposes.
		Time Complexity:
		Worst: O(n)  for reversed or random order
		Best: O(n)  when already sorted
		Space Complexity: O(1)  sorting in-place

```python
import time
import tracemalloc
import matplotlib.pyplot as plt
import random

# Bubble Sort algorithm
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Main function to test runtime and memory
def main():
    sizes = [100, 500, 1000, 2000, 3000]  # array sizes to test
    times = []
    memories = []

    for size in sizes:
        arr = [random.randint(1, 10000) for _ in range(size)]

        tracemalloc.start()
        start = time.time()

        bubble_sort(arr)

        end = time.time()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        times.append(end - start)
        memories.append(peak / 1024)  # in KB

        print(f"Array size: {size}, Time: {end - start:.4f}s, Memory: {peak/1024:.2f} KB")

    # Plotting
    plt.figure(figsize=(12, 5))

    # Runtime Plot
    plt.subplot(1, 2, 1)
    plt.plot(sizes, times, marker='o')
    plt.title("Bubble Sort Runtime")
    plt.xlabel("Array Size")
    plt.ylabel("Time (s)")

    # Memory Plot
    plt.subplot(1, 2, 2)
    plt.plot(sizes, memories, marker='o', color='orange')
    plt.title("Bubble Sort Memory Usage")
    plt.xlabel("Array Size")
    plt.ylabel("Memory (KB)")

    plt.tight_layout()
    plt.show()

    # Big-O Complexity
    print("\n Bubble Sort Time Complexity:")
    print("Worst-case: O(n^2)")
    print("Best-case (already sorted): O(n)")
    print("Space complexity: O(1) - in-place sort")

if __name__ == "__main__":
    main()
```

```python
import time
import tracemalloc
import matplotlib.pyplot as plt
import random

# Quick Sort Implementation
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]  # Choose pivot
    left = [x for x in arr if x < pivot]     # All smaller elements
    middle = [x for x in arr if x == pivot]  # All equal elements
    right = [x for x in arr if x > pivot]    # All larger elements
    return quick_sort(left) + middle + quick_sort(right)

# Function to measure performance
def measure_performance():
    sizes = [100, 500, 1000, 5000, 10000, 20000]
    runtimes = []
    memories = []

    for size in sizes:
        arr = [random.randint(0, 10000) for _ in range(size)]

        tracemalloc.start()
        start_time = time.time()

        quick_sort(arr)

        end_time = time.time()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        runtimes.append(end_time - start_time)
        memories.append(peak / 1024)  # Convert to KB

    # Plotting runtime
    plt.figure()
    plt.plot(sizes, runtimes, marker='o')
    plt.xlabel("Array Size")
    plt.ylabel("Runtime (seconds)")
    plt.title("Quick Sort Runtime Complexity")
    plt.grid(True)
    plt.show()

    # Plotting memory usage
    plt.figure()
    plt.plot(sizes, memories, marker='s')
    plt.xlabel("Array Size")
    plt.ylabel("Memory Usage (KB)")
    plt.title("Quick Sort Memory Usage")
    plt.grid(True)
    plt.show()

# Main entry
def main():
    print("Quick Sort Big-O Complexity:")
    print("Best Case: O(n log n)")
    print("Average Case: O(n log n)")
    print("Worst Case: O(n^2)\n")
    measure_performance()

if __name__ == "__main__":
    main()
```

```python
import time
import tracemalloc
import matplotlib.pyplot as plt

# Iterative Fibonacci function
def fibonacci(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Measure runtime and memory
def measure_performance():
    ns = [10, 100, 500, 1000, 5000, 10000]
    runtimes = []
    memories = []

    for n in ns:
        tracemalloc.start()
        start_time = time.time()

        fibonacci(n)

        end_time = time.time()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        runtimes.append(end_time - start_time)
        memories.append(peak / 1024)  # Convert to KB

    # Plot runtime
    plt.figure()
    plt.plot(ns, runtimes, marker='o')
    plt.title("Fibonacci Iterative: Runtime vs n")
    plt.xlabel("n")
    plt.ylabel("Runtime (seconds)")
    plt.grid(True)
    plt.show()

    # Plot memory
    plt.figure()
    plt.plot(ns, memories, marker='s')
    plt.title("Fibonacci Iterative: Memory vs n")
    plt.xlabel("n")
    plt.ylabel("Memory Usage (KB)")
    plt.grid(True)
    plt.show()

# Main
def main():
    print("Fibonacci Sequence (Iterative)")
    print("Big-O Time Complexity: O(n)")
    print("Big-O Space Complexity: O(1)\n")
    measure_performance()

if __name__ == "__main__":
    main()
```

		Big-O analysis: Time = O(2), Space = O(n) (due to call stack)
		Runtime + memory usage
		Plot to compare with iterative version

```python
import time
import tracemalloc
import matplotlib.pyplot as plt

# Recursive Fibonacci function (inefficient)
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Measure runtime and memory for small n (due to exponential time)
def measure_recursive_performance():
    ns = [5, 10, 15, 20, 25, 30]  # caution: n > 35 will be VERY slow
    runtimes = []
    memories = []

    for n in ns:
        tracemalloc.start()
        start_time = time.time()

        fibonacci_recursive(n)

        end_time = time.time()
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()

        runtimes.append(end_time - start_time)
        memories.append(peak / 1024)  # Convert to KB

    # Plot runtime
    plt.figure()
    plt.plot(ns, runtimes, marker='o', color='red')
    plt.title("Fibonacci Recursive: Runtime vs n")
    plt.xlabel("n")
    plt.ylabel("Runtime (seconds)")
    plt.grid(True)
    plt.show()

    # Plot memory
    plt.figure()
    plt.plot(ns, memories, marker='s', color='orange')
    plt.title("Fibonacci Recursive: Memory vs n")
    plt.xlabel("n")
    plt.ylabel("Memory Usage (KB)")
    plt.grid(True)
    plt.show()

# Main
def main():
    print("Fibonacci Sequence (Recursive)")
    print("Big-O Time Complexity: O(2^n)")
    print("Big-O Space Complexity: O(n) due to recursion depth\n")
    measure_recursive_performance()

if __name__ == "__main__":
    main()
```

|Method |Time  Complexity| Space Complexity| Notes |
|--------|--------|--------------|----------------|
|Recursive | O(2)| O(n) (due to call stack) | Elegant but not practicalfor large n|
|Iterative|O(n)|O(1)| Best for performance and memory efficiency|
